```python
# Question No #1
add = "+"
sub = "-"
mult = "*"
div = "/"

a = int(input("Enter 1st Value: "))
sel = input("Which operation you want (+, -, *, /): ")
b = int(input("Enter 2nd Value: "))

if sel == add:
    print("Addition of 2 values is", a + b)

elif sel == sub:
    print("Subtraction of 2 values is", a - b)

elif sel == mult:
    print("Multiplication of 2 values is", a * b)

elif sel == div:
    print("Division of 2 values is", a / b)

else:
    print("Invalid Operator")
```

```python
# Question NO #2
lst = ["apple", "banana","orange", 23, "shabbir"]
found = False

for item in lst:
    if type(item) == int:
        found = True
    else:
        if type(item) == float:
            found = True
        else:
            found = found  # no change

if found:
    print("Numeric value found: True")
else:
    print("Numeric value found: False")
```

```python
# Question No #4
dictionary = {
     "a": 100,
    "b": 200,
    "c": "Anas",
    "d": 300
}
total = 0
for value in dictionary.values():
    if type(value) == int:
        total = total + value
        print("The sum of integers in given dictionary is : ", total)
```

```python
# Question No #5
my_list = [1, 2, 3, 2, 4, 5, 1, 6]

duplicates = [] 

for item in my_list:
    if my_list.count(item) > 1:   # agar item ek se zyada baar apper hon
        if item not in duplicates:  # agar woh duplicates me na hon
            duplicates.append(item)

print("Duplicate values are:", duplicates)
```
